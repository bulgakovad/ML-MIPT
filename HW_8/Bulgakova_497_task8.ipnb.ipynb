**Вопрос 6**: На какой идее базируются алгоритмы аггломеративной кластеризации? Напишите формулу Ланса-Вильма

Агломеративная кластеризация это подвид иерархической кластеризации.
Сперва все кластеры одноэлементные. Затем происходит несколько итераций слияния, каждая из которых включает в себя нахождение двух ближайших кластера, слияние их в один, и вычисление расстояния от нового кластера до всех остальных. 

Формула Ланса-Вильма: 
Для определения расстояние $R(W,S)$ между кластерами $W = U\cup V$ и $S$, зная расстояния $R(U,S)$, $R(V,S)$, $R(U, V)$ используется следующая формула:

$R(U\cup V,S) = \alpha_U * R(U,S) + \alpha_V * R(V, S) + \beta * R(U, V) + \gamma * |R(U, S) - R(V, S)|$, 

$\alpha_U, \alpha_V, \beta, \gamma$ - числовые параметры

**Вопрос 8**: В чём отличия (основные упрощения) k-means от EM-алгоритма кластеризации?

EM использует нормальное распределение и исходя из этого подбирает параметры. k-means же проводит шаги гораздо проще, просто минимизирует функцию и не использует вероятностные идеи. Таким образом, k-means формирует шарообразные кластеры, а EM - кластеры любой формы.

** Вопрос 9 **Какой принцип работы графовых алгоритмов кластеризации?

В графовых алгоритмах выборка объектов представляется в виде графа, вершинам которого соответствуют объекты, а ребра имеют вес, равный расстоянию между объектами. Для кластеризации используется выделение связных компонент, построение минимального остовного дерева и послойная кластеризация.

** Задача 3 **
Докажите, что the k-means всегда сходится.

Если бы алгоритму подавалось бесконечное количество точек, то он бы постоянно пересчитывал центр и мог бы не сойтись. Но при конечном количестве точек центры в конечном итоге перестанут колебаться и будут выбраны окончательно, так как точки фиксированы.

Рассмотрим $\sum_{x_i \in X} \rho(x_i, c(x_i)), ~ c(x_i) - $ центр кластера, к которому относится $x_i$.

При передвижении центров кластеров данная сумма не возрастает, так как центр масс точек - точка, сумма квадратов расстояний от которой до остальных минимальна. При отнесении точек к ближайшим центрам кластеров данная сумма тоже не возрастает, так как точка может изменить кластер, если новый центр кластера ближе, что влечет уменьшение нашей суммы.
Кроме того наша сумма может принимать лишь конечное число значений (существует лишь конечное число разбиений точек на кластеры).

Таким образом, сумма убывает пока алгоритм не сойдется, и не не принимает дважды ни одного своего значения. А так как  возможных значений конечное количество, то алгоритм сойдется.

```python
import sys
sys.path.append('C:\Anaconda2\Lib\site-packages\scipy')

import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * F.dot(np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma) #посчитать SVD матрицы Sigma

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])
    
    m = np.argmax(var_i > 100 * varRetained) #вычислите необходимое число главных компонент
    
    print('%.2f %% variance retained in %d dimensions' % (var_i[m], m))
    #верните число m и точность, которая достигается при этом числе главных компонент

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:m] #только m главных компонент
    G = U_reduced.dot(F) #вычислить матрицу в преобразованном пространстве

    return G, U_reduced
```

```python
# Примените алгоритм к данным MNIST
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size=0.7)
```
